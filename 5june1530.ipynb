```python
for i in range(0, 1122):

    # Applying Regular Expression
    
    '''
    Replace email addresses with 'emailaddr'
    Replace URLs with 'httpaddr'
    Replace money symbols with 'moneysymb'
    Replace phone numbers with 'phonenumbr'
    Replace numbers with 'numbr'
    '''
    msg2 = df2['JD'][i]
    msg2 = re.sub('\b[\w\-.]+?@\w+?\.\w{2,4}\b', 'emailaddr', df2['JD'][i])
    msg2 = re.sub('(http[s]?\S+)|(\w+\.[A-Za-z]{2,4}\S*)', 'httpaddr', df2['JD'][i])
    msg2 = re.sub('Â£|\$', 'moneysymb', df2['JD'][i])
    msg2 = re.sub('\b(\+\d{1,2}\s)?\d?[\-(.]?\d{3}\)?[\s.-]?\d{3}[\s.-]?\d{4}\b', 'phonenumbr', df2['JD'][i])
    msg2 = re.sub('\d+(\.\d+)?', 'numbr', df2['JD'][i])
    
    ''' Remove all punctuations '''
    msg2 = re.sub('[^\w\d\s]', ' ', df2['JD'][i])
    
    if i<2:
        print("\t\t\t\t MESSAGE ", i)
    
    if i<2:
        print("\n After Regular Expression - Message ", i, " : ", msg2)
    
    # Each word to lower case
    msg2 = msg2.lower()    
    if i<2:
        print("\n Lower case Message ", i, " : ", msg2)
    
    # Splitting words to Tokenize
    msg2 = msg2.split()    
    if i<2:
        print("\n After Splitting - Message ", i, " : ", msg2)
    
    # Stemming with PorterStemmer handling Stop Words
    msg2 = [ps.stem(word) for word in msg2 if not word in set(stopwords.words('english'))]
    if i<2:
        print("\n After Stemming - Message ", i, " : ", msg2)
    
    # preparing Messages with Remaining Tokens
    msg2 = ' '.join(msg2)
    if i<2:
        print("\n Final Prepared - Message ", i, " : ", msg2, "\n\n")
    
    # Preparing WordVector Corpus
    corpus2.append(msg2)
```
